```python
#linear regression
#data
#x(week)                                     y(sales in thousand)
#--------------------------------------------------------------------
#1                                                      1.2
#2                                                      1.8
#3                                                      2.5
#4                                                      3.2
#5                                                      3.8
#linear regression formula --> y=a0+a1*x

#a1-->((mean(x*y))-(meanof(x)*meanof(y)))/meanof(x^2)-(mean of(x)^2)

#a0=mean(y) -a1*meanof(x)

 #         x            y            x^2           x*y
    
 #         1            1.2          1             1.2
 #         2            1.8          4             3.6
 #         3            2.5          9             7.5
 #         4            3.2          16            12.8
 #         5            3.8          25            19
#-------------------------------------------------------------
#sum:     15           12.5         55            44.1
#Average: x=3          y=2.5        11            8.88

#1.) x=3       y=2.5         meanof(x^2)=11        meanof(xy)=8.88
```

```python
import pandas as pd
df_3=pd.read_csv(r"C:\Users\naras\Downloads\Linear_regr_Salary_dataset.csv")
df_3
```

```python
df_3.shape
```

```python
df_3.isna().sum()
```

```python
x=df_3[['YearsExperience']]
y=df_3['Salary']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=101)
x_train,x_test,y_train,y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,np.round(y_pred))
acc
```
